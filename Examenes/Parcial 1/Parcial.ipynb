# **Punto 1**

```
import requests 
url = "https://listado.mercadolibre.com.co/guitarra-electrica#D[A:guitarra%20electrica]"
response = requests.get (url)
print (response)
```

```
import bs4
getDataSelect = lambda clase : bs4.BeautifulSoup(response.text, 'html.parser').select(clase)

dataTitle = getDataSelect('.ui-search-item__title')
dataPrice = getDataSelect('.price-tag-fraction')

getText = lambda elemento : elemento.text

itemNames = list(map(getText, dataTitle))
itemPrices = list(map(getText, dataPrice))

# Cada otro precio es lo que costaria en cuotas entones lo voy a quitar
del itemPrices[1::2]

print (itemNames)
print (itemPrices)
```

# **Punto 2**

```
nameLength = len(itemNames)
priceLength = len(itemPrices)

if (priceLength != nameLength):
  difference = abs(nameLength - priceLength)

  if (nameLength > priceLength):
    for i in range(difference):
      itemPrices.append("Sin Precio")
      
  else:
    for i in range(difference):
      itemNames.append("Sin Nombre")

guitarDict = {}
guitarDict["Names"] = itemNames
guitarDict["Prices"] = itemPrices
print(guitarDict)
```

# **Punto 3**

```
import pandas as pd

guitarDataFrame = pd.DataFrame(guitarDict)
guitarDataFrame.to_csv('Guitarras.csv', index = False)
```

# **Punto 4**

```
addText = lambda element: "Guitarra: " + element

modifiedNames = list(map(addText, itemNames))

modGuitarDict = {}
modGuitarDict["Names"] = modifiedNames
modGuitarDict["Prices"] = itemPrices

guitarModDataFrame = pd.DataFrame(modGuitarDict)
guitarModDataFrame.to_csv('GuitarrasMod.csv', index = False)
```

# **Punto 5**

```
brands = ["Squier", "Tagima", "Yamaha"]
functionList = []
filteredNames = []

for i in range(len(brands)):
  functionList.append(lambda element: brands[i] in element)
  filteredNames.append(list(filter(functionList[i], itemNames)))

filteredGuitarDict = {}
filteredGuitarDict["Brands"] = brands
filteredGuitarDict["Names"] = filteredNames

guitarBrandDataFrame = pd.DataFrame(filteredGuitarDict)
guitarBrandDataFrame.to_csv('GuitarrasMarcas.csv', index = False)
```

```
shortFilter = lambda element: len(element) < 80
shortNames = list(filter(shortFilter, itemNames))

guitarShortDataFrame = pd.DataFrame(shortNames)
guitarShortDataFrame.to_csv('GuitarrasNombreCorto.csv', index = False)
```

# **Punto 6**

```
length = lambda element: len(element)

nameLengths = list(map(length, itemNames))
average = sum(nameLengths) / len(itemNames)

print(average)
```
